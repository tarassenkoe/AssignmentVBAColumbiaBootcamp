```python
import pandas as pd
import numpy as np
```

```python
df1=pd.read_excel("Multiple_year_stock_data.xlsx")
df2=pd.read_excel("Multiple_year_stock_data.xlsx",sheet_name="2019")
df3=pd.read_excel("Multiple_year_stock_data.xlsx",sheet_name="2020")
```

```python
dffinal1234=pd.concat([df1,df2,df3],axis=0)
dffinal1234
```

```python
tickerdataframe['distinct_values']=pd.DataFrame(dffinal1234['<ticker>'].unique())
```

```python
df2final123=pd.merge(df1,tickerdataframe,how="inner",left_on="<ticker>",right_on="distinct_values")
```

```python
df2final123=df2final123.drop_duplicates(keep="first")
```

```python
df2final123['change']=df2final123['<close>']-df2final123['<open>']
```

```python
df2final123['percent_change']=(df2final123['<close>']-df2final123['<open>'])/df2final123['<close>']
```

```python
df2final123['stock_volume_at_close']=df2final123['<close>']*df2final123['<vol>']
```

```python
#BGI has the greatest change of the stocks.
df2final123.sort_values(by="percent_change",ascending=False)
```

```python
#BGI has the greatest change of the stocks.
df2final123.sort_values(by="percent_change",ascending=True)
```

```python
#BGI has the greatest change of the stocks.
df2final123.sort_values(by="percent_change",ascending=False)
```

```python
df2final123['stock_volume_at_close']=df2final123['<close>']*df2final123['<vol>']
```

```python
df2final1234=df2final123.sort_values(by="stock_volume_at_close",ascending=False)
```

```python
df2final1234.to_csv("Multiple_year_stock_data3.csv")
```
